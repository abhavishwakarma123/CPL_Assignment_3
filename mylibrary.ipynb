```python
##################################################################
'''This class performs all row operations'''
##################################################################
class Row_operations:
    
    def swap_rows(matrix, n, r, col):                  #swapping rows using traditional method
        temp = 0
        for j in range(col):
            temp = matrix[n][j]
            matrix[n][j] = matrix[r][j]
            matrix[r][j] = temp
        return matrix
    
    def multiply_row(matrix, r = [], fact = 0):        #multiplying a factor to the row
        row = []
        col = len(matrix[0])
        for x in range(col):
            row.append(fact*r[x])
        return row
    
    def add_rows(matrix, r1, r2):                      #adding two rows
        row = []
        col = len(matrix[0])
        for x in range(col):
            row.append(r1[x] + r2[x])
        return row
    
    def partial_pivot(matrix, p):                      #partial pivoting
        pivot = matrix[p][p]
        row = len(matrix)
        col = len(matrix[0])
        if pivot != 0:
            return matrix
        else:
            for i in range(p+1,row):
                if matrix[i][p] != 0:
                    Row_operations.swap_rows(matrix, i, p, col)
                    return matrix                      #will return as soon as a non zero pivot is
                                                       #found
```

```python
##################################################################
'''This class performs all matrix operations'''
##################################################################
class Matrix_operations:
    
    def augment_matrix(M_square, M_col):                  #augmenting two matrices
        M = []
        
        row = len(M_square)
        col = len(M_square[0])
        
        for i in range(row):
            M.append(M_square[i])

        for i in range(row):
            M[i] = np.append(M[i], M_col[i][0])           #np.append because np.arrays do not
                                                          #work with normal append function
        
        return M
            
    def make_proper_column(M):                            #makes a list a column matrix
        C = []                                            #of the form [[],[],[]...]    
        for i in M:
            temp = []
            temp.append(i)
            C.append(temp)
        return C
            
    def make_proper_row(M):                               #makes a list a row matrix
        R = []                                            #of the form [...]
        R.append(M)
        return R
    
    def mat_mult(X, Y):                                   #DEFINING THE MATRIX MULTIPLICATION FUNCTION
        len_row_X = len(X)
        len_column_X = len(X[0])
        len_row_Y = len(Y)
        len_column_Y = len(Y[0])

        if len_column_X == len_row_Y:                     #CHECKING IF MATRIX MULTIPLICATION IS POSSIBLE
            Z = []                                        #THIS WILL BE FINAL MULTIPLIED MATRIX
            for i in range(len(X)):
                Z_row_temp = []                           #THIS WILL BE THE TEMPORARY ROWS APPENDED TO Z AFTER EACH ITERATION
                for j in range(len(Y[0])):
                    temp = 0                              #THIS WILL BE THE TEMPORARY ELEMENTS OF EACH ROW
                    for k in range(len(Y)):
                        temp += X[i][k]*Y[k][j]           #THE MAIN ITERATION
                    Z_row_temp.append(temp)               #APPENDING THE ELEMENT TO THE ROW
                Z.append(Z_row_temp)                      #APPENDING THE ROW TO Z
            return Z
        else:
            print("Invalid Argument")
            
    def separate_aug_mat(M):                              #separating the augmented matrix
        A = []                                            #to its counterparts
        B = []
        row = len(M)
        
        for i in range(row):
            temp_A_row = []
            temp_B_row = []

            for j in range(2*row):
                if j < row:
                    temp_A_row.append(M[i][j])
                else:
                    temp_B_row.append(M[i][j])
            A.append(temp_A_row)
            B.append(temp_B_row)
            
        return A, B
                    
```

```python
##################################################################
'''This class performs rounding'''
##################################################################
class Round:                             

    def round_matrix(matrix):                                    #rounding each element of the matrix
        row = len(matrix)
        col = len(matrix[0])
        for i in range(row):
            for j in range(col):
                matrix[i][j] = Round.round_num(matrix[i][j])
        return matrix
    
    def round_num(num):                                          #rounding any number
        num_1 = num*100
        if num >= 0:
            num_1 = math.floor(num_1)
        else:
            num_1 = math.ceil(num_1)
        num = num_1*0.01
        return num
```

```python
##################################################################
'''This class performs all gauss jordan operations'''
##################################################################
class Gauss_Jordan:
    
    def solve(matrix):                                             #doing the complete gauss jordan
        row = len(matrix)                                          #elimination
        col = len(matrix[0])

        
        
        #looping for each pivot row
        
        for pivot_row in range(row):
            matrix = Row_operations.partial_pivot(matrix, pivot_row)     #partial pivoting pivot row

            
            #if even after pivoting pivot is zero then no solution
            
            if matrix[pivot_row][pivot_row] == 0:                        
                print("No solution exists")
                return

            else:
                pivot = matrix[pivot_row][pivot_row]
                pivot_inverse = 1/pivot
                
                #multiplying pivot row with inverse of pivot element to make the pivot element 1
                matrix[pivot_row] = Row_operations.multiply_row(matrix, matrix[pivot_row], pivot_inverse)
                
                
                #performing the row addition of each row except the pivot row, with the pivot row, 
                #after multiplying the pivot row with approporiate factor(say s) to make
                #every element in pivot column zero except the pivot element
                #   s = -1*matrix[m][pivot_row]
                
                
                for m in range(row):

                    #looping over each row and skipping pivot row
                    if m != pivot_row:
                        
                        #temp stores a copy of the pivot row multipled by s(defined in prev comments)
                        temp = Row_operations.multiply_row(matrix, matrix[pivot_row], -1*matrix[m][pivot_row])
                        
                        
                        #adding mth row with temp
                        matrix[m] = Row_operations.add_rows(matrix, matrix[m], temp)
        
        
        #rounding off the matrix elements
        matrix = Round.round_matrix(matrix)
        return matrix

    def determinant(matrix):                                #finds the determinant of matrix
        row = len(matrix)                                   #using row operations
        
        for pivot_row in range(row):
            
            for m in range(row):

                #performing the row addition of each row except the pivot row, with the pivot row, 
                #after multiplying the pivot row with approporiate factor(say s) to make
                #every element in pivot column zero except the pivot element
                #   s = -1*matrix[m][pivot_row]/matrix[pivot_row][pivot_row]
                
                
                if m != pivot_row:
                    temp = Row_operations.multiply_row(matrix, matrix[pivot_row], -1*(matrix[m][pivot_row])/(matrix[pivot_row][pivot_row]))                  
                    matrix[m] = Row_operations.add_rows(matrix, matrix[m], temp)

        matrix = Round.round_matrix(matrix)
        
        det = 1
        for i in range(row):
            det = det*matrix[i][i]
        
        return print("The determinant of this matrix is ", det)
        
```
